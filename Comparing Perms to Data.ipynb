```python
from permutations import *
import pandas as pd
from dateutil.relativedelta import relativedelta
```

```python
permutations=pd.read_csv('permutations.csv')
congress = pd.read_csv('congress_main.csv')
congress['TransactionDate'] = pd.to_datetime(congress['TransactionDate'])
returns = pd.read_csv('returns.csv')
returns.index = returns.Date.astype('datetime64[ns]')
returns.drop('Date', axis=1, inplace=True)
congress['Variance']=congress.apply(calculate_variance,axis=1)
congress.head()
```

```python
len(congress)*0.0005
```

```python
permutations=permutations[['Representative','Times Beat Var Rank']]
permutations=permutations.sort_values(by='Times Beat Var Rank')
print('Bottom 0.05 Percent:')
print(permutations.head(13))
print('Reps who always beat their variance:')
print(permutations[permutations['Times Beat Var Rank']==1000])
```

```python
ranks=congress.groupby('Representative')['Variance'].mean().sort_index(ascending=False)
ranks=pd.DataFrame(ranks).reset_index()
ranks=ranks.sort_values(by='Variance')
print("Congress's Bottom 0.05 percent Traders (Variance):")
print(ranks.head(13))
print("Congress's Top 0.05 percent Traders (Variance):")
print(ranks.tail(13))
```
